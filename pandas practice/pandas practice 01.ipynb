## Leader Names vs Years in Power (Bar Plot)

```python
import matplotlib.pyplot as plt

# create tenure column
df1["tenure"] = df1["end_year"] - df1["start_year"]

```

```python
plt.plot(plotsize=(10,6))
plt.bar(df1["name"], df1["tenure"])
plt.xticks(rotation=90)
plt.xlabel("Leaders")
plt.ylabel("Years in Power")
plt.title("Tenure Length of Pakistani Leaders")
plt.show()
```

## How Leaders Left Power (Bar Plot)

```python
plt.plot(plotsize=(8,5))
df1["how_left_power"].value_counts().plot(kind="bar")
plt.xlabel("How Leaders Left Power")
plt.ylabel("Number of Leaders")
plt.title("Modes of Exit from Power")
plt.show()
```

## Leader Names vs End Year

```python
plt.plot(plotsize=(10,6))
plt.bar(df1["name"], df1["end_year"])
plt.xticks(rotation=90)
plt.xlabel("Leaders")
plt.ylabel("End Year")
plt.title("End Year of Leadership by Leader")
plt.show()
```

## binning and also feature engineering


```python
import pandas as pd

year_bins = [1947, 1955, 1965, 1975, 1985, 1995, 2005, 2015, 2025]
year_labels = [
    "19471955",
    "19561965",
    "19661975",
    "19761985",
    "19861995",
    "19962005",
    "20062015",
    "20162025"
]

df1["start_year_bin"] = pd.cut(
    df1["start_year"],
    bins=year_bins,
    labels=year_labels,
    include_lowest=True
)

df1[["name", "start_year", "start_year_bin"]]
```

```python
df1.head()
```

## Binning tenure into year ranges (also YEARS)

```python
df1["tenure"] = df1["end_year"] - df1["start_year"]

tenure_bins = [0, 2, 4, 6, 10, 15]
tenure_labels = [
    "02 years",
    "34 years",
    "56 years",
    "710 years",
    "10+ years"
]

df1["tenure_bin"] = pd.cut(
    df1["tenure"],
    bins=tenure_bins,
    labels=tenure_labels,
    include_lowest=True
)

df1[["name", "tenure", "tenure_bin"]]
```
