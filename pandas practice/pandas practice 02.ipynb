`Description`
This dataset contains demographic information from the Pakistan Population Census conducted in 2017. It provides detailed population data at various administrative levels within Pakistan, including provinces, divisions, districts, and sub-divisions. The dataset also includes information on urban and rural populations, gender distribution, transgender individuals, sex ratios, population figures from the 1998 census, and annual growth rates.

`Features`
Province: The administrative provinces or regions of Pakistan where the census data was collected.

Division: The divisions within each province. Divisions are the second level of administrative divisions in Pakistan.

District: Districts within each division, representing larger administrative units.

Sub-Division: Sub-divisions or tehsils within each district, providing more localized data.

Area: The land area (in square kilometers) of each sub-division.

Urban Population 2017: The population of urban areas within each sub-division for the year 2017.

Rural Population 2017: The population of rural areas within each sub-division for the year 2017.

Male Population 2017: The male population within each sub-division for the year 2017.

Female Population 2017: The female population within each sub-division for the year 2017.

Transgender Population 2017: The population of transgender individuals within each sub-division for the year 2017.

Sex Ratio 2017: The sex ratio, calculated as the number of females per 1000 males, within each sub-division for the year 2017.

Population in 1998: The total population of each sub-division as recorded in the 1998 census.

Annual Growth Rate: The annual growth rate of the population in each sub-division, calculated as the percentage increase from 1998 to 2017.

# ASSIGNMENT: Calculate the percent change of Population?

```python
# ASSIGNMENT
# CALCULATE THE PERCENT CHANGE OF POPULATION FROM 1998 TO 2017 FOR URBAN AND RURAL AREAS
percent_change_urban = (pop_change_urban / pop_urban_1998) * 100
print("The Percent Change of Urban Population from 1998 to 2017 is:", percent_change_urban)
# CALCULATE THE SAME FOR RURAL POPULATION
pop_rural_1998 = df['POPULATION 1998 (RURAL)'].sum()
pop_rural_2017 = df['ALL SEXES (RURAL)'].sum()
pop_change_rural =  pop_rural_2017 - pop_rural_1998
percent_change_rural = (pop_change_rural / pop_rural_1998) * 100
print("The Percent Change of Rural Population from 1998 to 2017 is:", percent_change_rural) 
```

# **Assignment:** How to combine 3 columns in one column? make a separate data of everything based on Sex (Male, Female and Transgender?)

```python
# How to combine 3 columns in one column? make a separate data of everything based on Sex (Male, Female and Transgender?
df.head()
```

```python
#How to combine 3 columns in one column? make a separate data of everything based on Sex (Male, Female and Transgender?
df_male = df[['PROVINCE', 'DIVISION', 'DISTRICT', 'SUB DIVISION', 'AREA (sq.km)', 'POPULATION 1998 (URBAN)', 'POPULATION 1998 (RURAL)', 'MALE (URBAN)', 'MALE (RURAL)']]
df_female = df[['PROVINCE', 'DIVISION', 'DISTRICT', 'SUB DIVISION', 'AREA (sq.km)', 'POPULATION 1998 (URBAN)', 'POPULATION 1998 (RURAL)', 'FEMALE (URBAN)', 'FEMALE (RURAL)']]
df_transgender = df[['PROVINCE', 'DIVISION', 'DISTRICT', 'SUB DIVISION', 'AREA (sq.km)', 'POPULATION 1998 (URBAN)', 'POPULATION 1998 (RURAL)', 'TRANSGENDER (URBAN)', 'TRANSGENDER (RURAL)']]
df_male = df_male.rename(columns={'MALE (URBAN)': 'POPULATION (URBAN)', 'MALE (RURAL)': 'POPULATION (RURAL)'})
df_female = df_female.rename(columns={'FEMALE (URBAN)': 'POPULATION (URBAN)', 'FEMALE (RURAL)': 'POPULATION (RURAL)'})
df_transgender = df_transgender.rename(columns={'TRANSGENDER (URBAN)': 'POPULATION (URBAN)', 'TRANSGENDER (RURAL)': 'POPULATION (RURAL)'})
df_male['SEX'] = 'Male'
df_female['SEX'] = 'Female'
df_transgender['SEX'] = 'Transgender'
combined_df = pd.concat([df_male, df_female, df_transgender], ignore_index=True)
combined_df.head()
```

```python
df.columns
```

```python
#value.counts()
combined_df['SEX'].value_counts()
```

```python
#just the males in urban areas
male_urban = combined_df[(combined_df['SEX'] == 'Male') & (combined_df['POPULATION (URBAN)'] > 0)]
male_urban.head()  
```

```python
#value counts for males in urbnan areas
male_urban['SEX'].value_counts()   
```

```python
# valuecounts for  females in urban areas
female_urban = combined_df[(combined_df['SEX'] == 'Female') & (combined_df['POPULATION (URBAN)'] > 0)]
female_urban.head()
female_urban['SEX'].value_counts()
```

## just the percentage of each sex in urban areas like male, female, transgender

```python
total_urban_population = combined_df[combined_df['POPULATION (URBAN)'] > 0]['POPULATION (URBAN)'].sum()
male_urban_population = male_urban['POPULATION (URBAN)'].sum()
percentage_male_urban = (male_urban_population / total_urban_population) * 100
print("Percentage of Males in Urban Areas:", percentage_male_urban)
```

```python
female_urban_population = female_urban['POPULATION (URBAN)'].sum()
percentage_female_urban = (female_urban_population / total_urban_population) * 100
print("Percentage of Females in Urban Areas:", percentage_female_urban)  
```

```python
# what about transgender percentage?
transgender_urban = combined_df[(combined_df['SEX'] == 'Transgender') & (combined_df['POPULATION (URBAN)'] > 0)]
transgender_urban_population =  transgender_urban['POPULATION (URBAN)'].sum()
percentage_transgender_urban = (transgender_urban_population / total_urban_population) * 100
print("Percentage of Transgenders in Urban Areas:", percentage_transgender_urban)

```

```python
# need the bar plot for male, female,transgender,
sex_counts = combined_df[combined_df['POPULATION (URBAN)'] > 0]['SEX'].value_counts()
sex_counts.plot(kind='bar')
plt.title('Urban Population by Sex')            
plt.xlabel('Sex')
plt.ylabel('Number of Individuals')
plt.show()
```
