```python
# convert the whole size column into bytes

#let's define a function

def convert_size(size):
    if isinstance(size, str):
        if 'k' in size:
            return float(size.replace('k', "")) * 1024
        elif 'M' in size:
            return float(size.replace('M', "")) * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size
```

```python
dfff['Size']
```

```python
#applying the function to the Size column
dfff['Size'] = dfff['Size'].apply(convert_size)
```

```python
#checking the applied fuc
dfff['Size']
```

```python
# rename the column
dfff.rename(columns={'Size': "only_in_bytes"}, inplace=True)
```

```python
dfff.head()
```

```python
dfff['Size_in_Mb'] = dfff['only_in_bytes'].apply(lambda x: x/(1024*1024))
```

```python
dfff['Size_in_kb'] = dfff['only_in_bytes'].apply(lambda x: x/(1024))
```

# Installs Column:
1. Remove + sign
2. remove ,
3. Convert the column into an integer

```python
dfff['Installs'] = dfff['Installs'].apply(lambda x: x.replace('+', "") if '+' in str(x) else x)
```

```python
dfff['Installs'] = dfff['Installs'].apply(lambda x: x.replace(',', "") if ',' in str(x) else x)
```

```python
dfff['Installs'] = dfff['Installs'].apply(lambda x: int(x))
```

# now the price coloumn 

```python
dfff['Price'].value_counts()
```

```python
# let's see how many values have $ sign
dfff['Price'].loc[dfff['Price'].str.contains('\$')].value_counts().sum()
```

```python
# This code counts the number of values in the 'Price' column which contains 0 but does not contain $ sign
dfff['Price'].loc[(dfff['Price'].str.contains('0')) & (~dfff['Price'].str.contains('\$'))].value_counts().sum()
```

```python
dfff['Price'] = dfff['Price'].apply(lambda x: x.replace('$', "") if '$' in str(x) else x)
```

```python
dfff['Price'].value_counts()
```

```python
# now we can convert this into a numeric value
dfff['Price'] = dfff['Price'].apply(lambda x: float(x))
```

```python
dfff.describe()
```

```python
# using f string print min, max and average prices of the app
print(f"Min price is {dfff['Price'].min()}")
print(f"Max price is {dfff['Price'].max()}")
print(f"Average price is {dfff['Price'].mean()}")
```

# misssing values

```python
# find missing values
dfff.isnull().sum().sort_values(ascending=False)
```

```python
# find missing value percentage in th data
round(dfff.isnull().sum()/len(dfff)*100, 2).sort_values(ascending=False)
```

```python
# find total number of missing values
dfff.isnull().sum().sum()
```

```python
# plot missing values
plt.figure(figsize=(16, 6))
sns.heatmap(dfff.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# make figure size
plt.plot(plotsize=(16, 6))
# plot the null values by their percentage in each column
missing_percentage = dfff.isnull().sum()/len(dfff)*100
missing_percentage.plot(kind='bar')
# add the labels
plt.xlabel('Columns')
plt.ylabel('Percentage')
plt.title('Percentage of Missing Values in each Column')
```
