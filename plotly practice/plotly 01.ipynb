#  plotly data visulizatio 

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
names = ["Alice", "Bob", "Charlie"]
heights = [5.8, 5.9, 5.7]
# import library
import plotly.express as px
# create a plot (line chart)
fig = px.bar(x=names, y=heights, title="Height of Students") 
# show the fig
fig.show()
# save the plot in png file
fig.write_image("plots/01_barplot.png")

# save as standalone html
fig.write_html("html/01_barplot.html")
```

# plots on iris dataset

```python
# load the dataset
df = sns.load_dataset("iris")
df.head()
```

```python
# scatter plot as x scatter plot
fig = px.scatter(df, x="sepal_length", y="sepal_width", color="species", title="Iris Sepal Length vs Width")
fig.show()
# save the plot in png file
fig.write_image("plots/02_scatterplot.png") 
```

```python
import plotly.graph_objects as go
 #create simple line chart
fig = go.Figure(data=go.Scatter(x=[1, 2, 3], y=[4, 5, 6], mode='lines+markers'))
fig.show()
# save the plot in png file
fig.write_image("plots/03_linechart.png")   
```

```python
#customize and save the line chart
fig.update_layout(title="Simple Line Chart", xaxis_title="X Axis", yaxis_title="Y Axis")
fig.write_image("plots/04_custom_linechart.png")
fig.show()
```

```python
# heatsmap using plotly
data = np.random.rand(10,10)

heatmap = go.Heatmap(z=data)
fig = go.Figure(data=[heatmap])
fig.write_image("plots/05_heatmap.png") 
fig.show()
```

```python
#heatmap corr on iris dataset 
fig = px.imshow(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr())
fig.write_image("plots/06_iris_heatmap.png")
fig.show()
```
