```python
# CREATE A SUNBURST  USING PLOTLY ON DIAMOND DATASE
import plotly.express as px
import pandas as pd
import seaborn as sns
# load the diamonds dataset from seaborn
dff = sns.load_dataset('diamonds')
# create sunburst chart using plotly on loaded dataset
fig = px.sunburst(dff, path=['cut', 'color'], values='price', title='Sunburst Chart of Cut and Color by Price')

# save the plot in plots and html folder
fig.write_image("plots/sunburst_chart_cut_color_price.png")
fig.write_html("html/sunburst_chart_cut_color_price.html")
# show the plot
fig.show()


```

```python
# make sunburst chart on iris dataset
import plotly.express as px
import seaborn as sns
# load the iris dataset from seaborn
df_iris = sns.load_dataset('iris')
# create sunburst chart using plotly on loaded dataset
fig = px.sunburst(df_iris, path=['species', 'sepal_length'], title='Sunburst Chart of Species and Sepal Length')
# save the plot in plots and html folder
fig.write_image("plots/sunburst_chart_species_sepal_length.png")
fig.write_html("html/sunburst_chart_species_sepal_length.html")
# show the plot
fig.show()
```
